```sql
select
  sc.country,
  year(st.dateTime) as year,
  sum(st.totalPrice) as total_price
from
  sales_customers as sc
    inner join sales_transactions as st
      on st.customerID = sc.customerID
group by
  sc.country, year(st.dateTime);
```
