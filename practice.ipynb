# Creating Customers table and Orders table

```sql
DROP table IF EXISTS default.customers;
DROP table if exists default.orders;
```

```sql
CREATE TABLE default.customers (
  customer_id SERIAL PRIMARY KEY,
  full_name TEXT NOT NULL,
  email TEXT UNIQUE,
  city TEXT,
  signup_date DATE DEFAULT CURRENT_DATE,
  customer_type TEXT CHECK (customer_type in ('Regular', 'Premium', 'Enterprise'))
);

CREATE TABLE default.orders (
  order_id SERIAL PRIMARY KEY,
  customer_id INT,
  order_date TIMESTAMP DEFAULT CURRENT_TIMESTAMP,
  amount NUMERIC(10,2) CHECK (amount > 0),
  status TEXT CHECK (status in ('Pending', 'Shipped', 'Cancelled')),
  payment_method TEXT,
  metadata JSONB,
  tags TEXT[],
  CONSTRAINT fk_customer 
    FOREIGN KEY (customer_id)
    REFERENCES customers(customer_id)
)
```
